```python
import numpy as np
# Generate simulated temperature readings from two sensors.
# Sensor A: mean 25 C, std 3 C, 200 readings.
# Sensor B: mean 27 C, std 4.5 C, 200 readings.
# Also generate 200 timestamps uniformly from 0 to 10 seconds.
# Use np.random.default_rng with a seed = last 4 digits of your Drexel ID.

seed = 6412  # replace with your Drexel ID last 4 digits
rng = np.random.default_rng(seed)
n = 200

# timestamps uniformly in [0, 10] (no sort so order is random)
timestamp = rng.uniform(0.0, 10.0, size=n)

# two sensors with requested means and stddevs
sensor_a = rng.normal(loc=25.0, scale=3.0, size=n)
sensor_b = rng.normal(loc=27.0, scale=4.5, size=n)

# peek at first values
print('timestamp[:5]=', timestamp[:5])
print('sensor_a[:5]=', sensor_a[:5])
print('sensor_b[:5]=', sensor_b[:5])
```

```python
import matplotlib.pyplot as plt
# Scatter plot of sensor readings
plt.figure(figsize=(8,5))
plt.scatter(timestamp, sensor_a, s=20, alpha=0.7, label='Sensor A')
plt.scatter(timestamp, sensor_b, s=20, alpha=0.7, label='Sensor B')
plt.xlabel('Time (s)')
plt.ylabel('Temperature (°C)')
plt.title('Sensor temperature readings')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Scatter plot: sensor reading (y-axis) vs timestamp in seconds (x-axis).
# Color Sensor A points blue and Sensor B points orange.
# Add axis labels with units, a title, and a legend.
import matplotlib.pyplot as plt

plt.figure(figsize=(9,5))
plt.scatter(timestamp, sensor_a, s=30, c='tab:blue', alpha=0.8, label='Sensor A')
plt.scatter(timestamp, sensor_b, s=30, c='tab:orange', alpha=0.8, label='Sensor B')

plt.xlabel('Time (s)')
plt.ylabel('Temperature (°C)')
plt.title('Sensor temperature readings vs Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Histogram of sensor readings for Sensor A and Sensor B
import matplotlib.pyplot as plt
import numpy as np

# use 30 bins and shared range for fair comparison
bins = 30

plt.figure(figsize=(8,5))
plt.hist(sensor_a, bins=bins, color='tab:blue', alpha=0.6, label='Sensor A', edgecolor='black', linewidth=0.2)
plt.hist(sensor_b, bins=bins, color='tab:orange', alpha=0.6, label='Sensor B', edgecolor='black', linewidth=0.2)

# dashed vertical lines at means of each sensor
mean_a = np.mean(sensor_a)
mean_b = np.mean(sensor_b)
plt.axvline(mean_a, color='tab:blue', linestyle='--', linewidth=2, label='Mean A')
plt.axvline(mean_b, color='tab:orange', linestyle='--', linewidth=2, label='Mean B')

plt.xlabel('Temperature (°C)')
plt.ylabel('Count')
plt.title('Histogram of sensor temperature readings')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Side-by-side box plot comparing Sensor A and Sensor B distributions.
# Label x-axis with sensor names, y-axis with "Temperature (deg C)".
# Add a horizontal dashed line at the overall mean of both sensors combined.
```

```python
# Box plot comparing Sensor A and Sensor B
import matplotlib.pyplot as plt
import numpy as np

# Prepare data
data = [sensor_a, sensor_b]
labels = ['Sensor A', 'Sensor B']

plt.figure(figsize=(7,6))
# Create boxplot with notches and colored boxes
bp = plt.boxplot(data, labels=labels, patch_artist=True, notch=True,
                 flierprops=dict(marker='o', markerfacecolor='gray', markersize=5, alpha=0.6))

# Color the boxes
colors = ['tab:blue', 'tab:orange']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.6)

# Plot mean markers
means = [np.mean(d) for d in data]
plt.scatter([1,2], means, color=colors, marker='D', edgecolor='black', zorder=3, label='Mean')

# horizontal dashed line at overall mean across both sensors
overall_mean = np.mean(np.concatenate(data))
plt.axhline(overall_mean, color='gray', linestyle='--', linewidth=1.5, label='Overall mean')

plt.ylabel('Temperature (°C)')
plt.title('Box plot of sensor temperature readings')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
